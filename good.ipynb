```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import copy
```

```python
img_beforeblur = cv2.imread('20180916162824-2.jpg')
img_beforeblur = cv2.cvtColor(img_beforeblur, cv2.COLOR_BGR2RGB)
plt.imshow(img_beforeblur)
```

```python
img_blur_small = cv2.GaussianBlur(img_beforeblur, (5,5), 0)
```

```python
hsv = cv2.cvtColor(img_blur_small, cv2.COLOR_RGB2HSV)
```

```python
dst = cv2.inRange(hsv, (100, 70, 100), (130, 255, 255))
```

```python
cv2.imshow("dst", dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
ret, img_bi = cv2.threshold(dst, 100, 255, cv2.THRESH_BINARY)
cv2.imshow("img_bi", img_bi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
nLabels, labelImage, stats, centroids = cv2.connectedComponentsWithStats(img_bi)
```
